```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Standard Brownian Motion
def brownian_motion(N, T):
    dt = T / N
    t = np.linspace(0, T, N)
    dW = np.sqrt(dt) * np.random.normal(0, 1, N)
    W = np.cumsum(dW)
    return t, W

# Geometric Brownian Motion
def geometric_brownian_motion(N, T, mu, sigma, S0):
    dt = T / N
    t = np.linspace(0, T, N)
    dW = np.sqrt(dt) * np.random.normal(0, 1, N)
    W = np.cumsum(dW)
    S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)
    return t, S
```

```python
# Parameters
T = 1.0  # Total time
N = 1000  # Number of steps
dt = T / N  # Time step
mu = 0.1  # Drift for Geometric Brownian Motion
sigma = 0.2  # Volatility
S0 = 100  # Initial value for Geometric Brownian Motion



# Simulate
# np.random.seed(42)  # For reproducibility
t_bm, W = brownian_motion(N, T)
t_gbm, S = geometric_brownian_motion(N, T, mu, sigma, S0)

# Plot results
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t_bm, W, label='Brownian Motion')
plt.title('Standard Brownian Motion')
plt.xlabel('Time')
plt.ylabel('W(t)')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t_gbm, S, label='Geometric Brownian Motion', color='orange')
plt.title('Geometric Brownian Motion')
plt.xlabel('Time')
plt.ylabel('S(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig('brownian_motion_simulation.png')
```
