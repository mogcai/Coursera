```python
import random
import matplotlib.pyplot as plt
```

```python
def martingale_simulation(initial_balance, initial_bet, max_bets):
    balance=initial_balance
    bet=initial_bet
    history=[balance]

    for _ in range(max_bets):
        outcome=random.randint(0,1)
        
        if outcome==1:
            balance+=bet
            bet=initial_bet
        else:
            balance-=bet
            bet*=2

        history.append(balance)

        if balance<=0 or balance<bet:
            print('Bankrupt!')
            break

    plt.plot(history)
    plt.xlabel('Bet Number')
    plt.ylabel('Balance')
    plt.title('Martingale Simulation')

    return balance
```

```python
initial_balance=1000
initial_bet=10
max_bets=100

final_balance=martingale_simulation(initial_balance, initial_bet, max_bets)
print(f'Final balance: {final_balance}')
```
