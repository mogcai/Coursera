```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Function to compute points on the unit ball for a given p-norm
def unit_ball(p, num_points=1000):
    theta = np.linspace(0, 2 * np.pi, num_points)
    x = np.cos(theta)
    y = np.sin(theta)
    
    if p == 'inf':  # L-infinity norm
        r = 1 / np.maximum(np.abs(x), np.abs(y))
    else:  # Lp norm
        r = 1 / (np.abs(x)**p + np.abs(y)**p)**(1/p)
    
    x = r * x
    y = r * y
    return x, y
```

```python

# Compute unit balls for L1, L2, L3, and L-infinity norms
x1, y1 = unit_ball(1)
x2, y2 = unit_ball(2)
x3, y3 = unit_ball(3)
x_inf, y_inf = unit_ball('inf')

# Plot
plt.figure(figsize=(8, 8))
plt.plot(x1, y1, label='L1 Norm', linewidth=2)
plt.plot(x2, y2, label='L2 Norm', linewidth=2)
plt.plot(x3, y3, label='L3 Norm', linewidth=2)
plt.plot(x_inf, y_inf, label='L∞ Norm', linewidth=2)

# Customize plot
plt.axis('equal')  # Equal scaling on x and y axes
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Unit Balls for Different Vector Norms')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
