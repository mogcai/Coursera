```python
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def polya_urn(initial_red_balls, initial_blue_balls, num_draws):

    
    red_balls=initial_red_balls
    blue_balls=initial_blue_balls
    total_balls=red_balls+blue_balls
    history=[(red_balls, blue_balls, total_balls)]

    for _ in range(num_draws):
        # darw=random.randint(0,1) # 0 for red, 1 for blue
        # if darw==0:
        #     red_balls+=1
        # else:
        #     blue_balls+=1
        prob_red = red_balls / total_balls
        if random.random() < prob_red:
            red_balls += 1
        else:
            blue_balls += 1
        total_balls=red_balls+blue_balls
        history.append((red_balls, blue_balls, total_balls))

    return history
```

```python
initial_red_balls=1
initial_blue_balls=1
num_draws=10000
history=polya_urn(initial_red_balls, initial_blue_balls, num_draws)
```

```python
df=pd.DataFrame(history, columns=['Red Balls', 'Blue Balls', 'Total Balls'])
```

```python
df['Percentage_Red'] = df['Red Balls'] / df['Total Balls'] * 100
df['Percentage_Blue'] = df['Blue Balls'] / df['Total Balls'] * 100
df[['Percentage_Red', 'Percentage_Blue']].plot(title='Polya Urn Model: Percentage of Red and Blue Balls Over Time')
```
